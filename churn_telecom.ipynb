```python
dataset = pd.read_csv("churn.csv")
```

```python
X = dataset.iloc[:,1:-1]
Y = dataset.iloc[:,-1].values
```

```python
X = pd.get_dummies(X, drop_first = True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(max_depth=5, min_samples_split=4)

dt.fit(X_train,Y_train)

Y_pred_dt = dt.predict(X_test)
```

```python
cm_svc=confusion_matrix(Y_pred_dt,Y_test)
acc_svc=accuracy_score(Y_pred_dt,Y_test)
cr_svc=classification_report(Y_pred_dt,Y_test)
```

```python
from sklearn.tree.export import export_text
from termcolor import colored

def get_code(tree, feature_names):
        left      = tree.tree_.children_left
        right     = tree.tree_.children_right
        threshold = tree.tree_.threshold
        features  = [feature_names[i] for i in tree.tree_.feature]
        value = tree.tree_.value
        rule = ''
        def recurse(left, right, threshold, features, node, rule):
                if (threshold[node] != -2):
                    
                    if left[node] != -1:
                        rule = rule + ' and ' + features[node] + '<=' + str(np.round(threshold[node], 2)) + ' '
                        recurse (left, right, threshold, features,left[node], rule)
                    
                    if right[node] != -1:
                        end = rule.find(features[node])
                        
                        rule = rule[:end - len(' and ')]
                        
                        rule = rule + ' and ' + features[node] + '>' + str(np.round(threshold[node], 2)) + ' '
                        recurse (left, right, threshold, features,right[node], rule)
                else:
                    rule = rule[len(' and '):]
                    print(colored('if', 'magenta'), end= ' ')
                    rule = ' '.join(rule.split())
                    non_fraud, fraud = value[node].ravel()
                    
                    
                    words = rule.split('and')
                    
                    for i in range(len(words)):
                        print(colored(words[i], 'blue'), end=' ')
                        print(colored('and', 'grey'), end=' ')
                    
                    if non_fraud > fraud:
                        print(colored("then Loyal customer\n", 'green'))
                    else:
                        print(colored("then Churn customer \n", 'red'))
        recurse(left, right, threshold, features, 0, rule)
```

```python
get_code(dt, dataset.columns[:-1])
```

```python
from sklearn import tree
import graphviz
dot_data = tree.export_graphviz(dt, out_file=None, 
                     feature_names=X.columns,  
                     class_names=['Loyal', 'Churn'],  
                     filled=True, rounded=True,  
                     special_characters=True)  
graph = graphviz.Source(dot_data)  
graph
graph.render('dt.png', view=True) 
```
